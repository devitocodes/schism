# Convergence testing the 1st-order acoustic example

In this notebook, we will convergence test Schism's immersed boundary implementation against the previously presented exact solution.

```python
from convergence_1st_order import calculate_error

refinements = [0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55,
               0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95,
               1.0]
errs_max = []

for refinement in refinements:
    max_err = calculate_error(refinement)
    errs_max.append(max_err)
```

Plotting the convergence, we can see we have an approximately fourth-order trend, and a knee at a grid increment of around $\Delta x = 0.02$, after which the order of convergence drops off somewhat.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.loglog(np.array(refinements)*2*np.pi/240, errs_max, 'kx')
plt.loglog(np.array(refinements)*2*np.pi/240, errs_max, 'k')
plt.title("Convergence of the 4th-order scheme")
plt.xlabel("Grid increment")
plt.ylabel("Maximum error")
plt.savefig("convergence_plot_1st_order")
plt.show()
```

Finding the convergence gradient before the knee, we see that the scheme attains just short of 4th-order.

```python
grad = -np.polyfit(np.log10(np.array(refinements[:6])*2*np.pi/240),
                   np.log10(errs_max[:6]), 1)[0]
grad
```

```python
grad_knee = -np.polyfit(np.log10(np.array(refinements[6:])*2*np.pi/240),
                        np.log10(errs_max[6:]), 1)[0]
grad_knee
```
